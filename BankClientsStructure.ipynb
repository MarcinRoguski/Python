```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv('c://PythonScripts//BankChurners.csv', usecols = ["Education_Level"])

x = data.value_counts()
myLabels = data['Education_Level'].unique()

plt.pie(x, labels = myLabels, autopct="%.2f%%")
plt.title("Client's Education Level")
plt.show() 

```

```python
data = pd.read_csv('c://PythonScripts//BankChurners.csv', usecols = ["Customer_Age"])
data["group_by_age"] = pd.cut(
    x=data["Customer_Age"], 
    bins=[20, 40, 60, 80] 
    )
y = data['group_by_age'].value_counts()
a = ['20-40', '40-60', '60-80']
plt.pie(y, labels = a, autopct="%.2f%%")
plt.title("Client's Age")
plt.show
```
